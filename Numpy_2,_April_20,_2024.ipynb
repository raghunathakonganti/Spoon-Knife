# List Comprehension Questions for Practice


1. What is a list comprehension and how does it differ from a traditional for-loop in Python?
   - Solution: A list comprehension provides a more concise way to create lists. It consists of brackets containing an expression followed by a `for` clause. Compared to a traditional for-loop that might include multiple lines to create and append to a list, a list comprehension streamlines the process into a single line. For example, `[x**2 for x in range(10)]` is more concise than using a loop to append squares to an empty list.

2. How do you write a simple list comprehension to create a list of squares of the first 10 integers?
   - Solution: `[x**2 for x in range(1, 11)]`

3. Can list comprehensions include conditions? Write an example that creates a list of even numbers from 1 to 20 using a list comprehension.
   - Solution: `[x for x in range(1, 21) if x % 2 == 0]`

4. How can you use a list comprehension to flatten a matrix (a list of lists) into a single list?
   - Solution: `[item for sublist in matrix for item in sublist]`, where `matrix` is the list of lists.

5. Is it possible to use multiple if conditions in a list comprehension? Provide an example where you create a list of numbers which are divisible by both 2 and 5, from 1 to 100.
   - Solution: `[x for x in range(1, 101) if x % 2 == 0 and x % 5 == 0]`

6. Write a list comprehension to convert all strings in a list to uppercase. What happens if the list contains non-string elements?
   - Solution: `[s.upper() for s in list if isinstance(s, str)]`. If the list contains non-string elements, this comprehension will skip them to avoid errors.

7. How would you use a list comprehension to extract all the keys from a dictionary where the corresponding values are greater than 10?
   - Solution: `[key for key, value in dictionary.items() if value > 10]`, assuming `dictionary` is your dictionary object.

8. Can list comprehensions also include an else clause? Create a list comprehension that replaces all odd numbers with 'Odd' and even numbers with 'Even' in a list from 1 to 10.
   - Solution: `['Even' if x % 2 == 0 else 'Odd' for x in range(1, 11)]`

9. How do you apply a function to each element in a list using a list comprehension? For example, apply a function that multiplies each element by 2 to a list of numbers.
   - Solution: `[double(x) for x in numbers]`, where `double` is defined as `def double(x): return x * 2`.

10. What are the potential drawbacks of using list comprehensions, especially with regards to readability and performance?
   - Solution: List comprehensions can sometimes reduce readability, especially with complex conditions or multiple nested loops. Performance-wise, they are usually faster than equivalent for-loops, but they can use more memory if the resulting list is large. It's also easy to overuse them, which can lead to code that's hard to understand and maintain.

# Logical Operations

````python
import numpy as np

# Prices of items on your shopping list
prices = np.array([50, 45, 25, 20, 35])

budget = 30

# Check if there's at least one item you can afford
can_afford = np.any(prices <= budget) # like checking for OR condition

if can_afford:
    print("You can buy at least one item on your list!")
else:
    print("Sorry, nothing on your list fits your budget.")
````

````python
budget = 500

# Check if all the products are in my budget
np.all(prices <= budget)
````

````python
import numpy as np

# Chores status: 1 for done, 0 for not done
chores = np.array([1, 1, 1, 1, 0])

# Check if all chores are done
all_chores_done = np.all(chores == 1) # like AND condition

if all_chores_done:
    print("Great job! You've completed all your chores. Time to play!")
else:
    print("Finish all your chores before you can play.")
````
